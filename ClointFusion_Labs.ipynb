# **GUI Based Functions**

We have 6 functions which take different inputs from users.

```
# Use this function to get any random input (text / int / float) from user

# NOTE: By default the value is mandatory field.
ret_val = cf.gui_get_any_input_from_user('enter any number')
print(ret_val)

ret_val = cf.gui_get_any_input_from_user('enter your name',mandatory_field=False)
print(ret_val)

# if you are reading password or any sensitive information from user use additional parameter password=True
```

```
# Use this function to get any (valid) file path from user
file_path = cf.gui_get_any_file_from_user('input file','txt')
print(file_path)

img_file_path = cf.gui_get_any_file_from_user('input image file','png')
print(img_file_path)
```

```
# Use this function to accept folder path from user.

user_folder_path = cf.gui_get_folder_path_from_user('input folder')
print(user_folder_path)
```

```
# Use this function to get Yes or No value from user
yes_no = cf.gui_get_consent_from_user('Continue playing ?')
print(yes_no)
```

```
# Use this function to get user's choices from the given list. 

options_for_user_lst = ['Red','Green','Blue']

user_choise = cf.gui_get_dropdownlist_values_from_user('list of colors',options_for_user_lst)
print(user_choise)

# Set multi_select=False (True by default)
user_choise = cf.gui_get_dropdownlist_values_from_user('list of colors',options_for_user_lst,multi_select=False)
print(user_choise)
```
